## Introducción a la Construcción de un Grafo de Conocimiento con LLaMA Index

La construcción de un grafo de conocimiento es un proceso que permite organizar información en forma de entidades y relaciones, proporcionando una representación estructurada y semántica de los datos. Un grafo de conocimiento facilita la comprensión de conexiones complejas entre diferentes piezas de información, permitiendo realizar consultas avanzadas y obtener respuestas más precisas. En el contexto de LLaMA Index, el grafo de conocimiento se construye a partir de documentos de texto, extrayendo entidades y relaciones relevantes mediante técnicas de procesamiento de lenguaje natural (PLN).

### Proceso de Construcción del Grafo de Conocimiento

1. **Definición de Entidades y Relaciones**:
   - El primer paso en la construcción de un grafo de conocimiento es definir las **entidades** (como personas, lugares, organizaciones, etc.) y las **relaciones** (como "trabaja en", "es parte de", "fue diagnosticado con", etc.) que se desean extraer de los textos. Esta definición es crucial para orientar el proceso de extracción y asegurarse de que el grafo represente adecuadamente el dominio de interés.

2. **Preparación del Conjunto de Documentos**:
   - Se recolectan y organizan los documentos de texto que contienen la información que se desea estructurar en el grafo. Estos documentos pueden ser artículos, ensayos, investigaciones, o cualquier otra forma de texto que contenga datos relevantes.

3. **Extracción de Entidades y Relaciones**:
   - Utilizando herramientas avanzadas de PLN, LLaMA Index procesa los documentos para identificar y extraer entidades y las relaciones definidas previamente. Esta extracción se puede realizar mediante modelos preentrenados que reconocen patrones lingüísticos y semánticos específicos en el texto.

4. **Almacenamiento en el Grafo de Propiedades**:
   - Una vez extraídas, las entidades y relaciones se almacenan en un **grafo de propiedades**. Un grafo de propiedades es una estructura de datos que permite almacenar nodos (entidades) y aristas (relaciones) junto con propiedades adicionales que describen características de cada entidad o relación. En este caso, Neo4j es una base de datos de grafos comúnmente utilizada para almacenar y gestionar el grafo de propiedades.

5. **Construcción del Índice de Grafo**:
   - Se construye un **índice de grafo de conocimiento** (`PropertyGraphIndex`) en LLaMA Index a partir del conjunto de documentos y el almacén de grafos de propiedades. Este índice facilita la consulta eficiente y estructurada del grafo, permitiendo a los usuarios realizar preguntas complejas y obtener respuestas basadas en las relaciones y entidades del grafo.

6. **Consulta y Análisis**:
   - Con el grafo de conocimiento construido, se pueden realizar consultas para explorar las relaciones entre entidades, descubrir patrones ocultos, y realizar análisis avanzados. LLaMA Index proporciona herramientas para interactuar con el grafo y extraer conocimiento útil de manera dinámica y flexible.

### Beneficios del Grafo de Conocimiento

La construcción de un grafo de conocimiento utilizando LLaMA Index ofrece numerosos beneficios, incluyendo:
- **Estructuración de Información**: Transforma información no estructurada en datos estructurados que pueden ser fácilmente consultados y analizados.
- **Consulta Avanzada**: Permite realizar consultas semánticas y complejas sobre los datos, proporcionando respuestas más precisas y contextualmente relevantes.

En resumen, la construcción de un grafo de conocimiento con LLaMA Index es una poderosa herramienta para organizar y explorar grandes cantidades de información de manera eficiente, proporcionando una base sólida para la extracción de conocimiento y el análisis de datos.

### Proceso de Creación de un Extractor de Camino de LLM Basado en Esquemas para Datos Médicos

En este proceso, se configura un sistema para la extracción de entidades y relaciones médicas a partir de texto utilizando un modelo de lenguaje grande (LLM) llamado LLaMA.

1. **Importación de Bibliotecas y Definiciones**:
   - Se importan las bibliotecas necesarias, como `Literal` de `typing`, y módulos específicos de LLaMA Index como `Ollama` y `SchemaLLMPathExtractor`.
   - Se define un conjunto de entidades médicas (`entities`) y relaciones médicas (`relations`) utilizando el tipo `Literal`. Esto establece los tipos de entidades y relaciones que se pueden extraer del texto, como "PERSON", "DISEASE", "SYMPTOM", y relaciones como "DIAGNOSED_WITH", "TREATED_WITH".

2. **Definición del Esquema de Validación Médica**:
   - Se crea un esquema de validación (`medical_validation_schema`) que define cuáles entidades pueden tener determinadas relaciones. Por ejemplo, una entidad de tipo "PERSON" puede tener relaciones de "DIAGNOSIS" con "DISEASE" o "CONDITION".
   - Este esquema de validación es crucial para garantizar que las relaciones extraídas por el modelo sean lógicas y pertinentes dentro del contexto médico.

3. **Configuración del Extractor de Camino Basado en Esquemas**:
   - Se inicializa el `SchemaLLMPathExtractor`, una herramienta que utiliza el modelo de lenguaje grande para extraer datos estructurados basados en el esquema definido.
   - El extractor se configura con los siguientes parámetros:
     - **LLM (Ollama)**: Se utiliza el modelo `llama3.1` configurado con una temperatura de 0 para resultados más deterministas y un modo JSON para la salida estructurada. El `request_timeout` se establece en 3600 segundos para permitir solicitudes prolongadas.
     - **Entidades y Relaciones Posibles**: Se especifican las entidades y relaciones previamente definidas, que el modelo puede identificar y extraer del texto.
     - **Esquema de Validación**: Se proporciona el esquema de validación (`kg_validation_schema`) para guiar la extracción de datos y asegurar que se sigan las reglas médicas definidas.
     - **Modo Estricto**: Se establece el parámetro `strict` en `True` para que solo se extraigan relaciones y entidades que coincidan con el esquema de validación. Esto asegura que cualquier salida esté completamente alineada con las definiciones médicas proporcionadas, evitando valores o relaciones fuera del esquema.

4. **Aplicación del Extractor en Datos Médicos**:
   - Una vez configurado, el `SchemaLLMPathExtractor` se puede aplicar a texto médico para extraer entidades y relaciones que son válidas y significativas según el esquema definido. Esto es útil para tareas como la creación de gráficos de conocimiento médico, el análisis de textos clínicos, y la organización de información médica compleja.

Este proceso permite extraer información estructurada y validada del texto utilizando un modelo de lenguaje avanzado, facilitando la organización y el análisis de datos médicos en un formato útil y consistente.


````python
from typing import Literal
from llama_index.llms.ollama import Ollama
from llama_index.core.indices.property_graph import SchemaLLMPathExtractor
# best practice to use upper-case
from typing import Literal

# Define the medical entities
entities = Literal[
    "PERSON",           # Individual, patient, or healthcare professional
    "HEALTHCARE_FACILITY",  # Hospital, clinic, or any medical institution
    "HEALTHCARE_PROVIDER",  # Doctor, nurse, or healthcare worker
    "DISEASE",          # Any specific disease or medical condition
    "SYMPTOM",          # Clinical symptom related to diseases or conditions
    "MEDICATION",       # Any drug or medicine used for treatment
    "PROCEDURE",        # Medical procedure or surgery
    "TEST",             # Diagnostic tests, lab tests, or screenings
    "MEDICAL_DEVICE",   # Instruments or devices used in medical treatments or diagnostics
    "ORGANIZATION"      # General organizations in the medical field (e.g., pharmaceutical companies)
]

# Define the medical relationships
relations = Literal[
    "DIAGNOSED_WITH",    # Person diagnosed with a disease or condition
    "TREATED_WITH",      # Person or disease treated with a medication or procedure
    "WORKS_AT",          # Healthcare provider works at a healthcare facility
    "PERFORMS",          # Healthcare provider performs a procedure or test
    "PRESCRIBES",        # Healthcare provider prescribes a medication
    "ASSOCIATED_WITH",   # Disease associated with a symptom, or a test associated with a disease
    "HAS_SYMPTOM",       # Disease or condition has a particular symptom
    "USES",              # Medical procedure or healthcare provider uses a medical device
    "PART_OF",           # Part of an organization (e.g., a department within a hospital)
]


# define which entities can have which relations
medical_validation_schema = {
    "relationships": [
        {
            "PERSON": [
                {"DIAGNOSIS": ["DISEASE", "CONDITION"]},
                {"TREATMENT": ["MEDICATION", "PROCEDURE"]},
                {"ALLERGY": ["SUBSTANCE"]},
                {"VISIT": ["LOCATION"]},
                {"PROVIDER": ["HEALTHCARE_PROVIDER"]},
                {"FAMILY_HISTORY": ["DISEASE", "CONDITION"]},
            ]
        },
        {
            "HEALTHCARE_PROVIDER": [
                {"EMPLOYMENT": ["HOSPITAL", "CLINIC"]},
                {"TREATMENT": ["PERSON", "DISEASE", "CONDITION"]},
                {"PROCEDURE": ["PERSON", "CONDITION"]},
                {"PRESCRIPTION": ["MEDICATION"]},
            ]
        },
        {
            "DISEASE": [
                {"SYMPTOM": ["SYMPTOM"]},
                {"TREATMENT": ["MEDICATION", "PROCEDURE"]},
                {"CAUSE": ["CONDITION", "LIFESTYLE", "GENETIC"]},
            ]
        },
        {
            "CONDITION": [
                {"SYMPTOM": ["SYMPTOM"]},
                {"TREATMENT": ["MEDICATION", "PROCEDURE"]},
                {"CAUSE": ["DISEASE", "LIFESTYLE", "GENETIC"]},
            ]
        },
        {
            "MEDICATION": [
                {"PRESCRIBED_FOR": ["DISEASE", "CONDITION"]},
                {"ADMINISTERED_BY": ["HEALTHCARE_PROVIDER"]},
            ]
        },
        {
            "PROCEDURE": [
                {"PERFORMED_BY": ["HEALTHCARE_PROVIDER"]},
                {"FOR_CONDITION": ["DISEASE", "CONDITION"]},
            ]
        },
        {
            "SYMPTOM": [
                {"INDICATIVE_OF": ["DISEASE", "CONDITION"]},
            ]
        },
        {
            "SUBSTANCE": [
                {"CAUSES": ["ALLERGY"]},
                {"USED_IN": ["PROCEDURE"]},
            ]
        },
        {
            "LOCATION": [
                {"FACILITY": ["HOSPITAL", "CLINIC"]},
                {"VISITED_BY": ["PERSON"]},
            ]
        },
        {
            "EVENT": [
                {"INVOLVES": ["PERSON", "HEALTHCARE_PROVIDER", "PROCEDURE", "DISEASE", "MEDICATION"]},
            ]
        },
    ]
}

kg_extractor = SchemaLLMPathExtractor(
    llm=Ollama(model="llama3.1", temperature=0, json_mode=True, request_timeout=3600),
    possible_entities=entities,
    possible_relations=relations,
    kg_validation_schema=medical_validation_schema,
    # if false, allows for values outside of the schema
    # useful for using the schema as a suggestion
    strict=True,
)
````

### Configuración de un `PropertyGraphIndex` utilizando Neo4j en LLaMA Index

Se configura un índice de gráfico de propiedades (`PropertyGraphIndex`) utilizando Neo4j como almacenamiento de gráficos en la biblioteca LLaMA Index. Este índice permite almacenar y consultar datos estructurados en forma de grafos, lo que es especialmente útil para aplicaciones que requieren análisis avanzado de relaciones y entidades.

1. **Importación de Módulos Necesarios**:
   - Se importan `StorageContext` de `llama_index.core` y `Neo4jPropertyGraphStore` de `llama_index.graph_stores.neo4j`. Estos módulos proporcionan las herramientas necesarias para gestionar el almacenamiento y la indexación de datos en Neo4j.

2. **Configuración del Almacenamiento en Neo4j**:
   - Se crea una instancia de `Neo4jPropertyGraphStore`, que actúa como un almacén de gráficos basado en propiedades para Neo4j. Se utilizan las credenciales de conexión (nombre de usuario, contraseña y URI) previamente configuradas para establecer la conexión con la base de datos Neo4j.

3. **Creación del Índice de Gráfico de Propiedades (`PropertyGraphIndex`)**:
   - Se utiliza `PropertyGraphIndex.from_documents` para crear un índice de gráfico de propiedades a partir de un conjunto de documentos. Este índice permite extraer y almacenar entidades y relaciones de los documentos en el gráfico de propiedades de Neo4j.
   - **Parámetros Clave**:
     - `documents`: Conjunto de documentos a partir del cual se extraen las entidades y relaciones.
     - `kg_extractors`: Una lista de extractores de gráficos de conocimiento (como `kg_extractor`), que se encargan de identificar y extraer las entidades y relaciones definidas en los documentos.
     - `property_graph_store`: El almacén de gráficos de propiedades (`graph_store`) donde se almacenará el índice. En este caso, es una instancia de `Neo4jPropertyGraphStore` conectada a una base de datos Neo4j.
     - `show_progress`: Parámetro opcional que, si se establece en `True`, muestra el progreso del proceso de indexación.
     - `use_async`: Indica si el proceso de indexación se debe ejecutar de manera asíncrona. Para evitar problemas en algunos entornos, es recomendable establecer este parámetro en `False`.

4. **Ejecución del Proceso de Indexación**:
   - Este proceso puede tomar un tiempo considerable dependiendo del tamaño de los documentos y la complejidad de los datos. Durante la indexación, las entidades y relaciones extraídas se almacenan en el gráfico de propiedades en Neo4j, permitiendo consultas avanzadas y análisis posteriores.

Esta configuración permite aprovechar la potencia de Neo4j para almacenar y gestionar datos estructurados en forma de grafos, facilitando el análisis de relaciones complejas y la extracción de conocimiento a partir de documentos textuales.

### Creación de un Motor de Consulta y Ejecución de Consultas en LLaMA Index

Se configura un motor de consulta utilizando el índice previamente creado y se realiza una consulta para extraer información específica de los documentos indexados. Esta funcionalidad permite interactuar con el índice para obtener respuestas basadas en el contenido de los documentos.

1. **Creación del Motor de Consulta**:
   - Se utiliza el método `as_query_engine` del objeto `index` para crear un motor de consulta. Este motor permite realizar consultas sobre los datos indexados y obtener respuestas estructuradas.
   - **Parámetros del Motor de Consulta**:
     - `include_text=True`: Este parámetro indica que las respuestas del motor de consulta deben incluir fragmentos de texto originales de los documentos que son relevantes para la consulta.
     - `response_mode="tree_summarize"`: Este parámetro especifica que el motor de consulta debe utilizar el modo de resumen de árbol para estructurar la respuesta. En este modo, se genera un resumen jerárquico basado en la estructura del contenido de los documentos.

2. **Ejecución de la Consulta**:
   - **Definición de la Pregunta**: Se utiliza el método `query` del motor de consulta (`query_engine`) para ejecutar una consulta específica. En este caso, la pregunta es: "Tell me more about what the author worked on at Interleaf".
   - **Obtención de la Respuesta**: El motor de consulta procesa la pregunta, analiza los documentos indexados y devuelve una respuesta que incluye información relevante extraída de los textos. Esta respuesta está basada en la información que el autor proporcionó sobre su trabajo en Interleaf.

Este proceso permite a los usuarios interactuar de manera dinámica con el índice de documentos, facilitando la extracción de conocimiento y respuestas a preguntas específicas basadas en el contenido textural almacenado.


````python
query_engine = index.as_query_engine(
    include_text=True, response_mode="tree_summarize"
)
# Tell me a question
response = query_engine.query(
    "Tell me more about what the author worked on at Interleaf", # Write a question
)
````
