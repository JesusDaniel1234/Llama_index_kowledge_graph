## Introduccin a la Construccin de un Grafo de Conocimiento con LLaMA Index

La construccin de un grafo de conocimiento es un proceso que permite organizar informacin en forma de entidades y relaciones, proporcionando una representacin estructurada y semntica de los datos. Un grafo de conocimiento facilita la comprensin de conexiones complejas entre diferentes piezas de informacin, permitiendo realizar consultas avanzadas y obtener respuestas ms precisas. En el contexto de LLaMA Index, el grafo de conocimiento se construye a partir de documentos de texto, extrayendo entidades y relaciones relevantes mediante tcnicas de procesamiento de lenguaje natural (PLN).

### Proceso de Construccin del Grafo de Conocimiento

1. **Definicin de Entidades y Relaciones**:
   - El primer paso en la construccin de un grafo de conocimiento es definir las **entidades** (como personas, lugares, organizaciones, etc.) y las **relaciones** (como "trabaja en", "es parte de", "fue diagnosticado con", etc.) que se desean extraer de los textos. Esta definicin es crucial para orientar el proceso de extraccin y asegurarse de que el grafo represente adecuadamente el dominio de inters.

2. **Preparacin del Conjunto de Documentos**:
   - Se recolectan y organizan los documentos de texto que contienen la informacin que se desea estructurar en el grafo. Estos documentos pueden ser artculos, ensayos, investigaciones, o cualquier otra forma de texto que contenga datos relevantes.

3. **Extraccin de Entidades y Relaciones**:
   - Utilizando herramientas avanzadas de PLN, LLaMA Index procesa los documentos para identificar y extraer entidades y las relaciones definidas previamente. Esta extraccin se puede realizar mediante modelos preentrenados que reconocen patrones lingsticos y semnticos especficos en el texto.

4. **Almacenamiento en el Grafo de Propiedades**:
   - Una vez extradas, las entidades y relaciones se almacenan en un **grafo de propiedades**. Un grafo de propiedades es una estructura de datos que permite almacenar nodos (entidades) y aristas (relaciones) junto con propiedades adicionales que describen caractersticas de cada entidad o relacin. En este caso, Neo4j es una base de datos de grafos comnmente utilizada para almacenar y gestionar el grafo de propiedades.

5. **Construccin del ndice de Grafo**:
   - Se construye un **ndice de grafo de conocimiento** (`PropertyGraphIndex`) en LLaMA Index a partir del conjunto de documentos y el almacn de grafos de propiedades. Este ndice facilita la consulta eficiente y estructurada del grafo, permitiendo a los usuarios realizar preguntas complejas y obtener respuestas basadas en las relaciones y entidades del grafo.

6. **Consulta y Anlisis**:
   - Con el grafo de conocimiento construido, se pueden realizar consultas para explorar las relaciones entre entidades, descubrir patrones ocultos, y realizar anlisis avanzados. LLaMA Index proporciona herramientas para interactuar con el grafo y extraer conocimiento til de manera dinmica y flexible.

### Proceso de Creacin de un Extractor de Camino de LLM Basado en Esquemas para Datos Mdicos

En este proceso, se configura un sistema para la extraccin de entidades y relaciones mdicas a partir de texto utilizando un modelo de lenguaje grande (LLM) llamado LLaMA.

1. **Importacin de Bibliotecas y Definiciones**:
   - Se importan las bibliotecas necesarias, como `Literal` de `typing`, y mdulos especficos de LLaMA Index como `Ollama` y `SchemaLLMPathExtractor`.
   - Se define un conjunto de entidades mdicas (`entities`) y relaciones mdicas (`relations`) utilizando el tipo `Literal`. Esto establece los tipos de entidades y relaciones que se pueden extraer del texto, como "PERSON", "DISEASE", "SYMPTOM", y relaciones como "DIAGNOSED_WITH", "TREATED_WITH".

2. **Definicin del Esquema de Validacin Mdica**:
   - Se crea un esquema de validacin (`medical_validation_schema`) que define cules entidades pueden tener determinadas relaciones. Por ejemplo, una entidad de tipo "PERSON" puede tener relaciones de "DIAGNOSIS" con "DISEASE" o "CONDITION".
   - Este esquema de validacin es crucial para garantizar que las relaciones extradas por el modelo sean lgicas y pertinentes dentro del contexto mdico.

3. **Configuracin del Extractor de Camino Basado en Esquemas**:
   - Se inicializa el `SchemaLLMPathExtractor`, una herramienta que utiliza el modelo de lenguaje grande para extraer datos estructurados basados en el esquema definido.
   - El extractor se configura con los siguientes parmetros:
     - **LLM (Ollama)**: Se utiliza el modelo `llama3.1` configurado con una temperatura de 0 para resultados ms deterministas y un modo JSON para la salida estructurada. El `request_timeout` se establece en 3600 segundos para permitir solicitudes prolongadas.
     - **Entidades y Relaciones Posibles**: Se especifican las entidades y relaciones previamente definidas, que el modelo puede identificar y extraer del texto.
     - **Esquema de Validacin**: Se proporciona el esquema de validacin (`kg_validation_schema`) para guiar la extraccin de datos y asegurar que se sigan las reglas mdicas definidas.
     - **Modo Estricto**: Se establece el parmetro `strict` en `True` para que solo se extraigan relaciones y entidades que coincidan con el esquema de validacin. Esto asegura que cualquier salida est completamente alineada con las definiciones mdicas proporcionadas, evitando valores o relaciones fuera del esquema.

4. **Aplicacin del Extractor en Datos Mdicos**:
   - Una vez configurado, el `SchemaLLMPathExtractor` se puede aplicar a texto mdico para extraer entidades y relaciones que son vlidas y significativas segn el esquema definido. Esto es til para tareas como la creacin de grficos de conocimiento mdico, el anlisis de textos clnicos, y la organizacin de informacin mdica compleja.

Este proceso permite extraer informacin estructurada y validada del texto utilizando un modelo de lenguaje avanzado, facilitando la organizacin y el anlisis de datos mdicos en un formato til y consistente.


````python
from typing import Literal
from llama_index.llms.ollama import Ollama
from llama_index.core.indices.property_graph import SchemaLLMPathExtractor
# best practice to use upper-case
from typing import Literal

# Define the medical entities
entities = Literal[
    "PERSON",           # Individual, patient, or healthcare professional
    "HEALTHCARE_FACILITY",  # Hospital, clinic, or any medical institution
    "HEALTHCARE_PROVIDER",  # Doctor, nurse, or healthcare worker
    "DISEASE",          # Any specific disease or medical condition
    "SYMPTOM",          # Clinical symptom related to diseases or conditions
    "MEDICATION",       # Any drug or medicine used for treatment
    "PROCEDURE",        # Medical procedure or surgery
    "TEST",             # Diagnostic tests, lab tests, or screenings
    "MEDICAL_DEVICE",   # Instruments or devices used in medical treatments or diagnostics
    "ORGANIZATION"      # General organizations in the medical field (e.g., pharmaceutical companies)
]

# Define the medical relationships
relations = Literal[
    "DIAGNOSED_WITH",    # Person diagnosed with a disease or condition
    "TREATED_WITH",      # Person or disease treated with a medication or procedure
    "WORKS_AT",          # Healthcare provider works at a healthcare facility
    "PERFORMS",          # Healthcare provider performs a procedure or test
    "PRESCRIBES",        # Healthcare provider prescribes a medication
    "ASSOCIATED_WITH",   # Disease associated with a symptom, or a test associated with a disease
    "HAS_SYMPTOM",       # Disease or condition has a particular symptom
    "USES",              # Medical procedure or healthcare provider uses a medical device
    "PART_OF",           # Part of an organization (e.g., a department within a hospital)
]


# define which entities can have which relations
medical_validation_schema = {
    "relationships": [
        {
            "PERSON": [
                {"DIAGNOSIS": ["DISEASE", "CONDITION"]},
                {"TREATMENT": ["MEDICATION", "PROCEDURE"]},
                {"ALLERGY": ["SUBSTANCE"]},
                {"VISIT": ["LOCATION"]},
                {"PROVIDER": ["HEALTHCARE_PROVIDER"]},
                {"FAMILY_HISTORY": ["DISEASE", "CONDITION"]},
            ]
        },
        {
            "HEALTHCARE_PROVIDER": [
                {"EMPLOYMENT": ["HOSPITAL", "CLINIC"]},
                {"TREATMENT": ["PERSON", "DISEASE", "CONDITION"]},
                {"PROCEDURE": ["PERSON", "CONDITION"]},
                {"PRESCRIPTION": ["MEDICATION"]},
            ]
        },
        {
            "DISEASE": [
                {"SYMPTOM": ["SYMPTOM"]},
                {"TREATMENT": ["MEDICATION", "PROCEDURE"]},
                {"CAUSE": ["CONDITION", "LIFESTYLE", "GENETIC"]},
            ]
        },
        {
            "CONDITION": [
                {"SYMPTOM": ["SYMPTOM"]},
                {"TREATMENT": ["MEDICATION", "PROCEDURE"]},
                {"CAUSE": ["DISEASE", "LIFESTYLE", "GENETIC"]},
            ]
        },
        {
            "MEDICATION": [
                {"PRESCRIBED_FOR": ["DISEASE", "CONDITION"]},
                {"ADMINISTERED_BY": ["HEALTHCARE_PROVIDER"]},
            ]
        },
        {
            "PROCEDURE": [
                {"PERFORMED_BY": ["HEALTHCARE_PROVIDER"]},
                {"FOR_CONDITION": ["DISEASE", "CONDITION"]},
            ]
        },
        {
            "SYMPTOM": [
                {"INDICATIVE_OF": ["DISEASE", "CONDITION"]},
            ]
        },
        {
            "SUBSTANCE": [
                {"CAUSES": ["ALLERGY"]},
                {"USED_IN": ["PROCEDURE"]},
            ]
        },
        {
            "LOCATION": [
                {"FACILITY": ["HOSPITAL", "CLINIC"]},
                {"VISITED_BY": ["PERSON"]},
            ]
        },
        {
            "EVENT": [
                {"INVOLVES": ["PERSON", "HEALTHCARE_PROVIDER", "PROCEDURE", "DISEASE", "MEDICATION"]},
            ]
        },
    ]
}

kg_extractor = SchemaLLMPathExtractor(
    llm=Ollama(model="llama3.1", temperature=0, json_mode=True, request_timeout=3600),
    possible_entities=entities,
    possible_relations=relations,
    kg_validation_schema=medical_validation_schema,
    # if false, allows for values outside of the schema
    # useful for using the schema as a suggestion
    strict=True,
)
````

### Configuracin de un `PropertyGraphIndex` utilizando Neo4j en LLaMA Index

Este ndice permite almacenar y consultar datos estructurados en forma de grafos, lo que es especialmente til para aplicaciones que requieren anlisis avanzado de relaciones y entidades.
